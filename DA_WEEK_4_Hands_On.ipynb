```python
import sqlite3

# Create an in-memory SQLite database (temporary storage)
conn = sqlite3.connect(":memory:")
cursor = conn.cursor()
```

```python
# Create Customers Table
cursor.execute('''
CREATE TABLE Customers (
    CustomerID INTEGER PRIMARY KEY,
    Name TEXT,
    City TEXT,
    JoinDate DATE
)''')

# Create Menu Table
cursor.execute('''
CREATE TABLE Menu (
    DishID INTEGER PRIMARY KEY,
    DishName TEXT,
    Category TEXT,
    Price REAL
)''')

# Create Orders Table
cursor.execute('''
CREATE TABLE Orders (
    OrderID INTEGER PRIMARY KEY,
    CustomerID INTEGER,
    DishID INTEGER,
    OrderDate DATE,
    Quantity INTEGER,
    TotalPrice REAL,
    FOREIGN KEY (CustomerID) REFERENCES Customers(CustomerID),
    FOREIGN KEY (DishID) REFERENCES Menu(DishID)
)''')

# Create Employees Table
cursor.execute('''
CREATE TABLE Employees (
    EmployeeID INTEGER PRIMARY KEY,
    Name TEXT,
    Role TEXT,
    Salary REAL
)''')

conn.commit()  # Save changes
print("Tables Created Successfully!")
```

```python
import pandas as pd

# Function to execute and display SQL results
def execute_query(query):
    result = cursor.execute(query).fetchall()
    df = pd.DataFrame(result, columns=[desc[0] for desc in cursor.description])
    return df
```

```python
execute_query("SELECT * FROM Customers WHERE City = 'New York';")
```

```python
execute_query("SELECT * FROM Orders WHERE TotalPrice > 30;")
```
